# 4. Produtos Mais e Menos Vendidos

```python
# Loop para processar cada loja
for i, url in enumerate(urls):
    try:
        loja = pd.read_csv(url)
        loja['und'] = loja['Preço']  # Apenas usando 'Preço' como marcador de unidade
        quantidade_por_loja = loja.groupby('Produto')['und'].count().sort_values(ascending=False)

        print(f"\nQuantidade de produtos vendidos da loja {i+1}:")
        print(quantidade_por_loja)
        print("------------------------------------")

        # Quantidade máxima e mínima
        qtd_mais_vendido = quantidade_por_loja.max()
        qtd_menos_vendido = quantidade_por_loja.min()

        # Produtos empatados como mais vendidos
        produtos_mais_vendidos = quantidade_por_loja[quantidade_por_loja == qtd_mais_vendido].index.tolist()
        # Produtos empatados como menos vendidos
        produtos_menos_vendidos = quantidade_por_loja[quantidade_por_loja == qtd_menos_vendido].index.tolist()

        print(f"Produtos mais vendidos ({qtd_mais_vendido} unidades): {', '.join(produtos_mais_vendidos)}")
        print(f"Produtos menos vendidos ({qtd_menos_vendido} unidades): {', '.join(produtos_menos_vendidos)}")
        print("======================================================================")

    except Exception as e:
        print(f"Erro ao processar a loja {i+1}: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# URLs dos arquivos CSV
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]


## Grafico faturamento
# Lista com o nome das lojas e faturamentos
nomes_lojas = [f"Loja {i+1}" for i in range(len(urls))]
faturamentos = []

print("Faturamento das Lojas")
print("===================")
# Loop para processar cada loja
for i, url in enumerate(urls):
    try:
        loja = pd.read_csv(url)
        loja['Faturamento'] = loja['Preço'] + loja['Frete']
        total_faturamento = loja['Faturamento'].sum().round(2)
        faturamentos.append(total_faturamento)
    except Exception as e:
        print(f"Erro ao processar a {nomes_lojas[i]}: {e}")
        faturamentos.append(0)

# Criar gráfico de pizza
plt.figure(figsize=(8, 8))
cores = ['skyblue', 'lightcoral', 'lightgreen', 'gold']
explode = [0.05] * len(faturamentos)  # Destacar todas as fatias levemente

plt.pie(
    faturamentos,
    labels=nomes_lojas,
    autopct='%1.1f%%',
    startangle=90,
    colors=cores,
    explode=explode,
    shadow=True
)

plt.title('Distribuição do Faturamento Total por Loja')
plt.axis('equal')  # Deixa o gráfico como um círculo

# Salvar gráfico
output_dir = "graficos_lojas"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'faturamento_total_por_loja_pizza.png')
plt.savefig(output_path)
print(f"\nGráfico de pizza salvo em: {output_path}")

plt.show()

#####Grafico por categoria de produtos da loja
print("Faturamento por categorias de produtos das Lojas")
print("================================================")
# Nomes das lojas
nomes_lojas = [f"Loja {i+1}" for i in range(len(urls))]

# Dicionário para armazenar o faturamento por categoria por loja
dados_por_loja = {}
todas_categorias = set()

# Loop de leitura e cálculo
for i, url in enumerate(urls):
    try:
        loja = pd.read_csv(url)
        loja['Faturamento'] = loja['Preço'] + loja['Frete']
        faturamento = loja.groupby('Categoria do Produto')['Faturamento'].sum().round(2)
        dados_por_loja[nomes_lojas[i]] = faturamento
        todas_categorias.update(faturamento.index)
    except Exception as e:
        print(f"Erro ao processar a loja {i+1}: {e}")

# Consolidar em um DataFrame
todas_categorias = sorted(todas_categorias)
df_faturamento = pd.DataFrame(index=todas_categorias)

for loja, dados in dados_por_loja.items():
    df_faturamento[loja] = dados

# Preencher categorias ausentes com 0
df_faturamento = df_faturamento.fillna(0)

# Gráfico de barras agrupadas
plt.figure(figsize=(14, 7))
largura_barra = 0.2
x = range(len(df_faturamento.index))
cores = ['blue', 'orange', 'green', 'red']

for i, loja in enumerate(df_faturamento.columns):
    plt.bar(
        [pos + i * largura_barra for pos in x],
        df_faturamento[loja],
        width=largura_barra,
        label=loja,
        color=cores[i % len(cores)]
    )

plt.xlabel('Categoria do Produto')
plt.ylabel('Faturamento (R$)')
plt.title('Faturamento por Categoria de Produto nas 4 Lojas')
plt.xticks([pos + 1.5 * largura_barra for pos in x], df_faturamento.index, rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# Salvar o gráfico
output_dir = "graficos_lojas"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'faturamento_por_categoria_4_lojas.png')
plt.savefig(output_path)
print(f"\nGráfico salvo em: {output_path}")

plt.show()

print("Avaliação dos cliente de produtos comprados nas Lojas")
print("=====================================================")

nomes_lojas = [f"Loja {i+1}" for i in range(len(urls))]
avaliacoes_medias = []

# Processa cada loja
for i, url in enumerate(urls):
    try:
        loja = pd.read_csv(url)
        media_avaliacao = loja['Avaliação da compra'].mean().round(2)
        avaliacoes_medias.append(media_avaliacao)
        print(f"Avaliação média da {nomes_lojas[i]}: {media_avaliacao}")
    except Exception as e:
        print(f"Erro na {nomes_lojas[i]}: {e}")
        avaliacoes_medias.append(0)

# Gráfico de linhas
plt.figure(figsize=(8, 6))
plt.plot(nomes_lojas, avaliacoes_medias, marker='o', linestyle='-', color='blue', linewidth=2)
plt.title('Avaliação Média da Compra por Loja')
plt.xlabel('Loja')
plt.ylabel('Avaliação Média (0 a 5)')
plt.ylim(0, 5)
plt.grid(True)

# Anotar valores nos pontos
for i, nota in enumerate(avaliacoes_medias):
    plt.text(i, nota + 0.1, f'{nota:.2f}', ha='center')

# Salvar o gráfico
output_dir = "graficos_lojas"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'avaliacao_media_por_loja_linha.png')
plt.savefig(output_path)
print(f"\nGráfico de avaliação salvo em: {output_path}")

plt.show()
```
